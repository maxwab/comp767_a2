```python
ts = tiles.get_intervals(5, 10)
```

```python
plt.figure(figsize=(10, 10))
plt.fill([0, 1, 1, 0], [0, 0, 1, 1], alpha=.5)
plt.axis('equal')

cl = ['k', 'r', 'b', 'y', 'g']

for i, (a, b) in enumerate(ts):
    for e in a:
        plt.plot([np.min(b), np.max(b)], [e, e], c=cl[i], alpha=.2)
    for e in b:
        plt.plot([e, e], [np.min(a), np.max(a)], c=cl[i], alpha=.2)

for j in np.linspace(-1, 1, 3):
    obs = np.array([[j, -1, 5]])
    code = tiles.encode(obs, ts)
    coord = angles.prep(obs).squeeze()
    plt.text(coord[0], coord[1], code)
    plt.scatter(coord[0], coord[1], marker='x')
for j in np.linspace(1, -1, 3):
    obs = np.array([[j, 1, 5]])
    code = tiles.encode(obs, ts)
    coord = angles.prep(obs).squeeze()
    plt.text(coord[0], coord[1], code)
    plt.scatter(coord[0], coord[1], marker='x')
```
