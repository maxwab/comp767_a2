```python
import matplotlib                         
import matplotlib.pyplot as plt           
import seaborn as sns                     
sns.set()                                 
```

```python
import argparse as ap
from pathlib import Path
```

```python
import numpy as np
```

```python
lambda_ = 0.9
```

```python
alphas = [1/4, 1/8, 1/16]
seeds = list(range(10))
```

```python
plt.figure()
```

```python
p = Path('log/expe1/l:{}'.format(lambda_))
for a in alphas:
    phis = []
    for s in seeds:
        filename = 'phi_l:{}_a:{}_s:{}.npy'.format(lambda_, a, s)
        phi = np.load(p / filename)
        phis.append(phi)
    phis = np.stack(phis, axis=1)
    phis_m = phis.mean(1)
    plt.plot(phis_m, label='alpha = {}'.format(a))
plt.legend()
plt.show()
```
