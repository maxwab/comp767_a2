```python
%matplotlib inline
import numpy as np
import tiles
import matplotlib.pyplot as plt
```

```python
z = tiles.get_intervals(5, 10)
```

```python
plt.figure(figsize=(10, 10))
plt.fill([0, 1, 1, 0], [0, 0, 1, 1], alpha=.5)
plt.axis('equal')

cl = ['k', 'r', 'b', 'y', 'g']

for i, (a, b) in enumerate(z):
    for e in a:
        plt.plot([np.min(b), np.max(b)], [e, e], c=cl[i])
    for e in b:
        plt.plot([e, e], [np.min(a), np.max(a)], c=cl[i])
```
